Step 1: Import libraries & Load the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('/content/heart.csv')
```

Step 3: Basic info

```python
print("First 5 rows of the dataset:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())
```

Step 4: Check for missing values

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

Step 5: Histograms for all numeric columns

```python
df.hist(figsize=(14, 10), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Features", fontsize=16)
plt.tight_layout()
plt.show()
```

Step 6: Boxplots to detect outliers

```python
plt.figure(figsize=(16, 8))
for i, column in enumerate(df.columns[:9], 1):
    plt.subplot(3, 3, i)
    sns.boxplot(y=df[column], color='lightcoral')
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()
```

Step 7: Correlation matrix

```python
plt.figure(figsize=(12, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

Step 8: Pairplot (can be heavy for large datasets)


```python
sns.pairplot(df, hue='target')
plt.suptitle("Pairplot of Features Colored by Target", y=1.02)
plt.show()
```

Step 9: Skewness check

```python
print("\nSkewness of Numeric Features:")
print(df.skew(numeric_only=True))
```
